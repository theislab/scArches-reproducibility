```python
import os
import seaborn as sns
import scanpy as sc
import matplotlib.pyplot as plt
import json
import numpy as np
```

```python
sc.settings.set_figure_params(dpi=100, frameon=False)
```

```python
model = 'tranvae_1000'
tests = [1,2,3,4]
version = ''
dir_path = os.path.expanduser(f'~/Documents/benchmarking_results/rqr/{model}/pbmc/')
    
p_b = sns.color_palette("husl", 9)
batch_color = p_b
p_c = sns.color_palette("hls", 16)
ct_color = p_c
```

```python
for test in tests:
    save_path = os.path.expanduser(f'~/Documents/master_thesis_figures/tranvae_rqr/pbmc/umaps/')
    if not os.path.exists(save_path):
        os.makedirs(save_path)
    latent = sc.read(f'{dir_path}test_{test}/query_unsupervised/full_data.h5ad')
    if test == 1:
        latent.obs.batch = latent.obs.batch.replace('Freytag', 'Freytag (query)')
        latent.obs.batch = latent.obs.batch.replace('Oetjen_A', 'Oetjen_A (query)')
        latent.obs.batch = latent.obs.batch.replace('Oetjen_P', 'Oetjen_P (query)')
        latent.obs.batch = latent.obs.batch.replace('Oetjen_U', 'Oetjen_U (query)')
        latent.obs.batch = latent.obs.batch.replace('Sun_sample1_CS', 'Sun_sample1_CS (query)')
        latent.obs.batch = latent.obs.batch.replace('Sun_sample2_KC', 'Sun_sample2_KC (query)')
        latent.obs.batch = latent.obs.batch.replace('Sun_sample3_TB', 'Sun_sample3_TB (query)')
        latent.obs.batch = latent.obs.batch.replace('Sun_sample4_TC', 'Sun_sample4_TC (query)')
    if test == 2:
        latent.obs.batch = latent.obs.batch.replace('Freytag', 'Freytag (query)')
        latent.obs.batch = latent.obs.batch.replace('Sun_sample1_CS', 'Sun_sample1_CS (query)')
        latent.obs.batch = latent.obs.batch.replace('Sun_sample2_KC', 'Sun_sample2_KC (query)')
        latent.obs.batch = latent.obs.batch.replace('Sun_sample3_TB', 'Sun_sample3_TB (query)')
        latent.obs.batch = latent.obs.batch.replace('Sun_sample4_TC', 'Sun_sample4_TC (query)')
    if test == 3:
        latent.obs.batch = latent.obs.batch.replace('Freytag', 'Freytag (query)')
        
    if test == 1:
        sc.pl.umap(
            latent,
            color="batch",
            frameon=False,
            ncols=1,
            show=False,
            palette=batch_color
        )
        plt.savefig(f"{save_path}{test}_batch.pdf", bbox_inches='tight')
        sc.pl.umap(
            latent,
            color="celltype",
            frameon=False,
            ncols=1,
            show=False,
            palette=ct_color
        )
        plt.savefig(f"{save_path}{test}_celltype.pdf", bbox_inches='tight')
    else:
        sc.pl.umap(
            latent,
            color="batch",
            title='',
            frameon=False,
            ncols=1,
            show=False,
            palette=batch_color
        )
        plt.savefig(f"{save_path}{test}_batch.pdf", bbox_inches='tight')
        sc.pl.umap(
            latent,
            color="celltype",
            title='',
            frameon=False,
            ncols=1,
            show=False,
            palette=ct_color
        )
        plt.savefig(f"{save_path}{test}_celltype.pdf", bbox_inches='tight')
    
```
