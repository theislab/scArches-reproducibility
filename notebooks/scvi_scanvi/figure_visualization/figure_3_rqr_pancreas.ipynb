```python
import os
import seaborn as sns
import scanpy as sc
import matplotlib.pyplot as plt
import json
import numpy as np
```

```python
sc.settings.set_figure_params(dpi=100, frameon=False)
```

```python
model = 'tranvae_1000'
tests = [1,2,3,4,5]
version = ''
dir_path = os.path.expanduser(f'~/Documents/benchmarking_results/rqr/{model}/pancreas/')
    
b_c = sc.pl.palettes.godsnot_102[10:15]
batch_color = [b_c[2], b_c[3], b_c[0],b_c[4], b_c[1]]
ct_c = sc.pl.palettes.godsnot_102
ct_color = [ct_c[0], ct_c[2],ct_c[3],ct_c[4],ct_c[5],ct_c[6],ct_c[7],ct_c[1]]
f2_b = [batch_color[0],batch_color[4],batch_color[1],batch_color[3],batch_color[2]]
f2_c = [ct_color[0],ct_color[7],ct_color[1],ct_color[2],ct_color[3],ct_color[4],ct_color[5],ct_color[6]]
```

```python
for test in tests:
    save_path = os.path.expanduser(f'~/Documents/master_thesis_figures/tranvae_rqr/pancreas/umaps/')
    if not os.path.exists(save_path):
        os.makedirs(save_path)
    latent = sc.read(f'{dir_path}test_{test}/query_unsupervised/full_data.h5ad')
    if test == 1:
        latent.obs.batch = latent.obs.batch.replace('Pancreas CelSeq', 'Pancreas CelSeq (query)')
        latent.obs.batch = latent.obs.batch.replace('Pancreas CelSeq2', 'Pancreas CelSeq2 (query)')
        latent.obs.batch = latent.obs.batch.replace('Pancreas Fluidigm C1', 'Pancreas Fluidigm C1 (query)')
        latent.obs.batch = latent.obs.batch.replace('Pancreas SS2', 'Pancreas SS2 (query)')
    if test == 2:
        latent.obs.batch = latent.obs.batch.replace('Pancreas CelSeq', 'Pancreas CelSeq (query)')
        latent.obs.batch = latent.obs.batch.replace('Pancreas CelSeq2', 'Pancreas CelSeq2 (query)')
        latent.obs.batch = latent.obs.batch.replace('Pancreas Fluidigm C1', 'Pancreas Fluidigm C1 (query)')
    if test == 3:
        latent.obs.batch = latent.obs.batch.replace('Pancreas CelSeq', 'Pancreas CelSeq (query)')
        latent.obs.batch = latent.obs.batch.replace('Pancreas Fluidigm C1', 'Pancreas Fluidigm C1 (query)')
    if test == 4:
        latent.obs.batch = latent.obs.batch.replace('Pancreas Fluidigm C1', 'Pancreas Fluidigm C1 (query)')
        
    if test == 1:
        sc.pl.umap(
            latent,
            color="batch",
            frameon=False,
            ncols=1,
            show=False,
            palette=f2_b
        )
        plt.savefig(f"{save_path}{test}_batch.pdf", bbox_inches='tight')
        sc.pl.umap(
            latent,
            color="celltype",
            frameon=False,
            ncols=1,
            show=False,
            palette=f2_c
        )
        plt.savefig(f"{save_path}{test}_celltype.pdf", bbox_inches='tight')
    else:
        sc.pl.umap(
            latent,
            color="batch",
            title='',
            frameon=False,
            ncols=1,
            show=False,
            palette=f2_b
        )
        plt.savefig(f"{save_path}{test}_batch.pdf", bbox_inches='tight')
        sc.pl.umap(
            latent,
            color="celltype",
            title='',
            frameon=False,
            ncols=1,
            show=False,
            palette=f2_c
        )
        plt.savefig(f"{save_path}{test}_celltype.pdf", bbox_inches='tight')
    
```
