```python
import os
import seaborn as sns
import scanpy as sc
import matplotlib.pyplot as plt
import json
import numpy as np
```

```python
sc.settings.set_figure_params(dpi=100, frameon=False)
```

```python
model = 'scanvi'
freezing = 'freezed_expr'
dir_path = os.path.expanduser(f'~/Documents/benchmarking_results/figure_2/{model}/first_cond/{freezing}/')
save_path = os.path.expanduser(f'~/Documents/master_thesis_figures/figure_2/{model}_first/{freezing}/')
if not os.path.exists(save_path):
    os.makedirs(save_path)
```

```python
f1 = sc.read(f'{dir_path}full_data.h5ad')
f1.obs.batch = f1.obs.batch.replace('Tabula_muris', 'Tabula_muris (query)')
f1.obs.batch = f1.obs.batch.replace('Zeisel', 'Zeisel (query)')
f1
```

```python
b_c = sc.pl.palettes.godsnot_102
batch_color = b_c
ct_c = sns.color_palette("husl", 10)
ct_color = ct_c
```

```python
sc.pl.umap(
    f1,
    color="batch",
    frameon=False,
    ncols=1,
    show=False,
    palette=batch_color
)
plt.savefig(f"{save_path}ref_batch.pdf", bbox_inches='tight')
sc.pl.umap(
    f1,
    color="celltype",
    frameon=False,
    ncols=1,
    show=False,
    palette=ct_color
)
plt.savefig(f"{save_path}ref_ct.pdf", bbox_inches='tight')
```
