```python
import os
import seaborn as sns
import scanpy as sc
import matplotlib.pyplot as plt
import json
import numpy as np
```

```python
sc.settings.set_figure_params(dpi=100, frameon=False)
```

```python
model = 'tranvae_1000'
tests = [1,2,3,4]
version = ''
dir_path = os.path.expanduser(f'~/Documents/benchmarking_results/rqr/{model}/brain/')
    
b_c = sc.pl.palettes.godsnot_102
batch_color = b_c
ct_c = sns.color_palette("husl", 10)
ct_color = ct_c
```

```python
for test in tests:
    save_path = os.path.expanduser(f'~/Documents/master_thesis_figures/tranvae_rqr/brain/umaps/')
    if not os.path.exists(save_path):
        os.makedirs(save_path)
    latent = sc.read(f'{dir_path}test_{test}/query_unsupervised/full_data.h5ad')
    if test == 1:
        latent.obs.batch = latent.obs.batch.replace('Tabula_muris', 'Tabula_muris (query)')
        latent.obs.batch = latent.obs.batch.replace('Zeisel', 'Zeisel (query)')
        latent.obs.batch = latent.obs.batch.replace('Saunders', 'Saunders (query)')
    if test == 2:
        latent.obs.batch = latent.obs.batch.replace('Tabula_muris', 'Tabula_muris (query)')
        latent.obs.batch = latent.obs.batch.replace('Zeisel', 'Zeisel (query)')
    if test == 3:
        latent.obs.batch = latent.obs.batch.replace('Tabula_muris', 'Tabula_muris (query)')
        
    if test == 1:
        sc.pl.umap(
            latent,
            color="batch",
            frameon=False,
            ncols=1,
            show=False,
            palette=batch_color
        )
        plt.savefig(f"{save_path}{test}_batch.pdf", bbox_inches='tight')
        sc.pl.umap(
            latent,
            color="celltype",
            frameon=False,
            ncols=1,
            show=False,
            palette=ct_color
        )
        plt.savefig(f"{save_path}{test}_celltype.pdf", bbox_inches='tight')
    else:
        sc.pl.umap(
            latent,
            color="batch",
            title='',
            frameon=False,
            ncols=1,
            show=False,
            palette=batch_color
        )
        plt.savefig(f"{save_path}{test}_batch.pdf", bbox_inches='tight')
        sc.pl.umap(
            latent,
            color="celltype",
            title='',
            frameon=False,
            ncols=1,
            show=False,
            palette=ct_color
        )
        plt.savefig(f"{save_path}{test}_celltype.pdf", bbox_inches='tight')
    
```
