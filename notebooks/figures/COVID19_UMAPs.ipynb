```python
import scanpy as sc
import seaborn as sns
```

```python
path_to_save = "./results/integration/covid19/"
sc.settings.figdir = path_to_save
sc.settings.set_figure_params(dpi=100)
```

```python
latent_adata = sc.read("/home/mohsen/data/covid_lisa/latent_integrated.h5ad")
latent_adata
```

```python
label_key = 'study'
palette = sns.color_palette('husl', 13)
sc.pl.umap(latent_adata, color=label_key, palette=palette, show=True, frameon=False)
sc.pl.umap(latent_adata, color=label_key, palette=palette, show=False,
           legend_loc=None, frameon=False, title='',
           save=f'_integrated_{label_key}.pdf')
```

```python
label_key = 'celltype'
palette = sc.pl.palettes.godsnot_102
sc.pl.umap(latent_adata, color=label_key, palette=palette, show=True, frameon=False)
sc.pl.umap(latent_adata, color=label_key, palette=palette,
           show=False, legend_loc=None, frameon=False,  title='',
           save=f'_integrated_{label_key}.pdf')
```

```python
label_key = 'condition'
palette = [sns.color_palette('husl', 3)[1]] + [sns.color_palette('husl', 3)[2]] + [sns.color_palette('husl', 3)[0]]
sc.pl.umap(latent_adata, color=label_key, palette=palette, show=True, frameon=False)
sc.pl.umap(latent_adata, color=label_key, palette=palette, 
           show=False, legend_loc=None, frameon=False, title='',
           save=f'_integrated_{label_key}.pdf')
```
